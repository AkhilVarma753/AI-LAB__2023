```python
import numpy as np
import matplotlib.pyplot as plt


def plot_samples(X, Y, w=None):
    for i, sample in enumerate(X):
        if Y[i] == -1:
            plt.scatter(sample[0], sample[1], s=120, marker='_', linewidths=2)
        else:
            plt.scatter(sample[0], sample[1], s=120, marker='+', linewidths=2)

    if w is not None:
        x_vals = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
        y_vals = - (w[0] / w[1]) * x_vals - (w[2] / w[1])
        plt.plot(x_vals, y_vals, 'g-', label='Hyperplane')

    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    plt.show()


def perceptron_algorithm(X, Y, plot_errors=False):
    w = np.zeros(len(X[0]))
    eta = 1
    n = 30
    errors = []

    for t in range(n):
        total_error = 0
        for i, x in enumerate(X):
            if (np.dot(X[i], w) * Y[i]) <= 0:
                total_error += (np.dot(X[i], w) * Y[i])
                w = w + eta * X[i] * Y[i]

        if plot_errors:
            errors.append(total_error * -1)

    if plot_errors:
        plt.plot(errors)
        plt.xlabel('Epoch')
        plt.ylabel('Total Loss')
        plt.show()

    return w



# Toy dataset
X = np.array([
    [-2, 4, -1],
    [4, 1, -1],
    [1, 6, -1],
    [2, 4, -1],
    [6, 2, -1],
])


y = np.array([-1, -1, 1, -1, -1])

w = perceptron_algorithm(X, y, plot_errors=True)
plot_samples(X, y, w)
```
